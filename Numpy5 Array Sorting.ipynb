```python
import numpy as np
```

```python
arr=np.array([[9,3,2],
              [1,6,5],
              [2,8,7]])
```

```python
np.sort(arr,axis=0,kind="mergesort")
# axis=0 means sort columnWise
```

```python
np.sort(arr,axis=1,kind="mergesort")
# axis=1 means sort rowWise
```

```python
np.sort(arr,kind="mergesort")
# if axis is not provided, it'll automatically sort it rowWise
```
