```python
import numpy as np
```

```python
arr1=np.array([[1,2,3,4],[5,6,7,8]])
arr2=np.array([[9,8,7,6],[5,4,3,2]])
```

```python
arr1
```

```python
arr2
```

```python
# 2 functions for merging
# vstack() ....vertical stack ....for merging row wise
# hstack() ....horizontal stack ....for merging row wise

np.vstack((arr1,arr2))
```

```python
np.vstack((arr2,arr1))
```

```python
np.hstack((arr1,arr2))
```

```python
np.hstack((arr2,arr1))
```

```python
np.concatenate((arr1,arr2),axis=1)
# axis=0 means vertical stack
```

```python
np.concatenate((arr1,arr2),axis=1)
# axis=1 means horizontal stack
```

```python
# split is the reverse process of stack
# hsplit()....hstack()
# vsplit()....vstack()

np.hsplit(arr1,2)  # slpit into 2 parts
```

```python
np.vsplit(arr1,2)  # slpit into 2 parts
```
